```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data1 = pd.read_excel('Public-Dataset-Daily-County-Cases-17-25-Years.xlsx')
data1.head(10)
```

```python
data_max= pd.DataFrame(data1.groupby(by='COUNTY').max())
data_max.reset_index(inplace=True)
data_max.head()
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(26, 50))
sns.barplot(y=data_max.COUNTY,x=data_max.TOTAL_CASES)
for i in ax.patches:
    plt.text(i.get_width()+0.7, i.get_y()+0.6,
             str(round((i.get_width()), 2)),
             fontsize=20, fontweight='bold',
             color='blue',alpha=0.5)
# Add Plot Title
ax.set_title('TN 5-18Years Total Cases By County',
             loc='CENTER',color="blue" )
  
# Add Text watermark
fig.text(0.2, 0.0, 'Osamah Neamah', fontsize=14,
         color='green', ha='left', va='bottom',
         alpha=0.6)
  
# Show Plot
plt.show()
```

```python
data2= data_max.nlargest(10,['TOTAL_CASES'])
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(26, 13))
sns.barplot(y=data2.COUNTY,x=data2.TOTAL_CASES)
for i in ax.patches:
    plt.text(i.get_width()+0.7, i.get_y()+0.4,
             str(round((i.get_width()), 2)),
             fontsize=20, fontweight='bold',
             color='blue',alpha=0.5)
# Add Plot Title
ax.set_title('TN TOP 10 CONUTIES 17-25 Years',
             loc='CENTER',color="blue" )
  
# Add Text watermark
fig.text(0.2, 0.0, 'Osamah Neamah', fontsize=14,
         color='green', ha='left', va='bottom',
         alpha=0.6)
  
# Show Plot
plt.show()
```
